```python

def q_non_breaking_2022(Hm0, Rc, sm, berm_level, B, gamma_f=0.44):

    term1 = ((sm * B)/Hm0)**1.3
    term2 = (berm_level/(sm*Rc))**0.2
    
    # Bereken gamma_b
    gamma_b = 1 - 18 * term1 * (1 - 0.34 * term2)
    
    # --- FIX: BEGRENZEN GAMMA_B ---
    # Voorkom dat gamma_b negatief wordt. Een veilige ondergrens is 0.6.
    #gamma_b = np.maximum(0.6, gamma_b_raw)
    
    Rc_Hs = Rc / Hm0
    
    # ... (rest van je code voor exponent en q_star) ...
    numerator = 2.4 * Rc
    denominator = gamma_f * gamma_b * Hm0
    
    exponent = - (numerator / denominator)
    
    # (cot^-1 alpha) is interpreted as 1/cot(alpha) = tan(alpha)
    q_star = 0.016 * (1/sm) * np.exp(exponent)
    
    return q_star, gamma_b, Rc_Hs

# --- TOEPASSEN OP DE DATAFRAME ---

# 1. Definieer de constanten die niet in je dataframe staan
# VUL HIER JOUW EIGEN WAARDES IN! (np.radians zet graden om naar de benodigde radialen)          

# 2. Voer de functie uit met de kolommen uit df_master
# Omdat de functie 3 dingen returnt, vangen we die op in 3 variabelen
verwachte_q, berekende_gamma, berekende_rc_hs = q_non_breaking_2022(
    Hm0 = df_master['HS'], 
    Rc = df_master['Rc'],
    sm = df_master['steepness'], 
    berm_level = df_master['BL'], 
    B = df_master['BW']
)

# 3. Voeg de output toe als nieuwe kolommen aan je dataframe
df_master['q_star_verwacht_2022'] = verwachte_q

#####

#2025

#####

def overtopping_non_breaking_2025(
    Hm0, Rc, sm10, B, BL,
    cot_alpha =1.33, tan_alpha = 0.75, Hm0_swell=0.0, gamma_v=1.0, g=9.81, gamma_f=0.44
):
    if Hm0 <= 0 or sm10 <= 0: return 0.0, 0.0
    
    # 2. Gamma_b
    if cot_alpha <= 4.0:
        term1 = (sm10 * B) / Hm0
        term2 = BL / (sm10 * Rc)
        gamma_b = 1.0 - 18.0 * term1**1.3 * (1.0 - 0.34 * term2**0.2)
    else:
        gamma_b = 1.0
    

    # 3. Hoofdformule
    xi_m10 = tan_alpha / np.sqrt(sm10)
    denominator = (gamma_f * gamma_b * gamma_v * xi_m10**0.24 * Hm0)
    
    if denominator == 0: return 0.0, 0.0

    exponent = -2.5 * (Rc - 0.4 * Hm0_swell) / denominator
    exponent = np.clip(exponent, -100, 100)

    q_star = 0.8 * tan_alpha * np.exp(exponent)
    q = q_star * np.sqrt(g * Hm0**3)
    
    return q, q_star


# 2. FIX: Gebruik een hulpfunctie voor de apply methode!
# Deze functie krijgt van Pandas één specifieke rij (row) tegelijk binnen.
def bereken_2025_rij(row):
    q_star_berekend = overtopping_non_breaking_2025(
        Hm0 = row['HS'],         # LET OP: We gebruiken row['...'] in plaats van df_master['...']
        Rc = row['Rc'], 
        sm10 = row['steepness'], 
        B = row['BW'], 
        BL = row['BL']
    )
    return q_star_berekend

# 3. Voer het uit op de gehele DataFrame!
# Door axis=1 te gebruiken, dwingen we Pandas om de functie rij-voor-rij aan te roepen.
# Omdat de functie nu steeds maar 1 getal per keer ziet, falen de if-statements niet meer.
df_master['q_star_verwacht_2025'] = df_master.apply(bereken_2025_rij, axis=1)


# 1. Verwijder de ongewenste kolom
if 'q_star_verwacht_2' in df_master.columns:
    df_master = df_master.drop(columns=['q_star_verwacht_2'])

# 2. Maak een slimme functie die alles uitpakt (list, tuple, array, series)
def pak_getal_uit(x):
    # Zolang het een 'verpakking' is, pakken we het eerste element [0]
    if isinstance(x, (list, tuple, np.ndarray, pd.Series)):
        if len(x) > 0:
            return pak_getal_uit(x[0]) # We doen dit recursief voor dubbele haken zoals [[0.5]]
        else:
            return np.nan
    return x

# 3. Pas de functie toe op de kolom
df_master['q_star_verwacht_2025'] = df_master['q_star_verwacht_2025'].apply(pak_getal_uit)

# 4. Converteer nu veilig naar floats
df_master['q_star_verwacht_2025'] = df_master['q_star_verwacht_2025'].astype(float)

# Bekijk het resultaat

display(df_master.loc[df_master['steepness'].idxmax()])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_overtopping(df, condities_om_te_plotten):
    # 1. Filter de data op basis van de lijst die je meegeeft
    df_gefilterd = df[df['Run_ID'].isin(condities_om_te_plotten)]
    
    # Check even of je geen typfout hebt gemaakt in de namen
    if df_gefilterd.empty:
        print(f"Let op: Geen data gevonden voor de condities: {condities_om_te_plotten}")
        return
    
    # 2. Maak de plot met de GEFILTERDE data
    plt.figure(figsize=(10, 6))
    
    sns.scatterplot(
        data=df_gefilterd, 
        x='Rc/Hm0', 
        y='q_star', 
        hue='Conditie',    
        style='Herhaling', 
        s=120,             
        # We laten seaborn zelf de kleuren kiezen, tenzij je een vaste 'palette' forceert
    )
    
    plt.yscale('log')
    plt.title('Wave Overtopping: $q^*$ vs $R_c / H_{m0}$', fontsize=14)
    plt.xlabel('$R_c / H_{m0}$ [-]', fontsize=12)
    plt.ylabel('$q^*$ [-]', fontsize=12)
    plt.grid(True, which="both", ls="--", alpha=0.5)
    plt.legend(loc='lower left')
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_verwacht(df, runs_om_te_plotten=None):
    # 1. Filter nu specifiek op de exacte Run_ID (de herhaling)
    if runs_om_te_plotten is not None:
        df_plot = df[df['Run_ID'].isin(runs_om_te_plotten)].copy()
    else:
        df_plot = df.copy() # Teken alles als je de haakjes leeg laat
        
    if df_plot.empty:
        print(f"Let op: Geen data gevonden voor de opgegeven runs!")
        return
        
    plt.figure(figsize=(11, 7))
    
    # 2. Plot de GEMETEN data (stippen/symbolen)
    sns.scatterplot(
        data=df_plot,
        x='Rc/Hm0',
        y='q_star',
        hue='Conditie',    # Kleur blijft per hoofdconditie
        style='Herhaling', # Symbool per herhaling (R1, R2, etc.)
        s=120,
        palette='tab10',
        alpha=0.9
    )
    
    # 3. Plot de VERWACHTE data (kruisjes)
    sns.scatterplot(
        data=df_plot,
        x='Rc/Hm0',
        y='q_star_verwacht_2025',
        hue='Conditie',
        marker='X',        
        s=100,             
        palette='tab10',
        legend=False,      
        alpha=0.6          
    )

    plt.scatter([], [], color='grey', marker='X', s=100, label='Verwacht 2025')

    sns.scatterplot(
        data=df_plot,
        x='Rc/Hm0',
        y='q_star_verwacht_2022',
        hue='Conditie',
        marker='v',        
        s=100,             
        palette='tab10',
        legend=False,      
        alpha=0.6          
    )
    
    # Voeg het kruisje toe aan de legenda voor de duidelijkheid
    plt.scatter([], [], color='grey', marker='v', s=100, label='Verwacht 2022')
    
    # Opmaak
    plt.yscale('log')
    plt.title('Wave Overtopping: Gemeten vs Verwacht', fontsize=14)
    plt.xlabel('$R_c / H_{m0}$ [-]', fontsize=12)
    plt.ylabel('$q^*$ [-]', fontsize=12)
    
    plt.grid(True, which="both", ls="--", alpha=0.5)
    plt.legend(loc= 'lower left')
    plt.tight_layout()
    plt.show()


# plot_verwacht(df_master, ['SWL_B1_2P_R3'])
# plot_verwacht(df_master, ['EM_B1_2P_R1'])
plot_verwacht(df_master, ['EM_B1_2P_R1', 'EM_B1_4P_R1'])
plot_verwacht(df_master, ['REF_2P', 'EM_B1_2P_R1', 'EM_B2_2P_R1', 'SWL_B1_2P_R1', 'SWL_B2_2P_R1'])
```
